# Classificatin par arbre de décision
Comme deusième approche de classification, nous utiliserons l'arbre de décision.

1-A partir du module tree du package sklearn, importer la classe DecisionTreeClassifier.

2- Créer une instance de la classe `DecisionTreeClassifier`.

3- Exécuter l'instance sur les données d'apprentissage `x_train` afin de construire votre premier modèle prédictif que vous appellerez « modèle_DT ».  Pour ce faire, il faut exécuter la fonction fit de la classe DecisionTreeClassifier qui prend comme argument les ensembles les données explicatives `x_train` et les données cibles `y_train`.

````
##Importations

#importer la classe decisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

#importer la fonction train_test_split 
from sklearn.model_selection import train_test_split 

#importer scikit-learn metrics module pour le calcul de l'accuracy 
from sklearn import metrics 

#importer la fonction export_graphviz pour afficher l'arbre dans le notebook Jupyter
from sklearn.tree import export_graphviz

#importer pydotplus
import pydotplus

from sklearn.externals.six import StringIO  
from IPython.display import Image  

col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("/content/drive/MyDrive/ML/data/pima.csv", header=None, names=col_names)
pima.head()

feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
x = pima[feature_cols] 
y = pima.label

# 70% training and 30% test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) 

##Création du modèle d'arbre de décision

#création d'une instance de la classe
modele_DT = DecisionTreeClassifier()

modele_DT = modele_DT.fit(x_train,y_train)

y_pred = modele_DT.predict(x_test)

## Evaluation du modèle
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# bon accuracy 

## Visualisation des arbres de décision

dot_data = StringIO()
export_graphviz(modele_DT, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('pima.png')
Image(graph.create_png())




````

# Classificatin par la forêt d'arbres
Comme troisième approche de classification, nous utiliserons un forêt d'arbre de décision.

1-A partir du module `ensemble` du package sklearn, importer la classe `RandomForestClassifier`.

2- Créer une instance de la classe `RandomForestClassifier`.

3- Exécuter l'instance sur les données d'apprentissage `x_train` afin de construire votre premier modèle prédictif que vous appellerez « modèle_RF ».  Pour ce faire, il faut exécuter la fonction fit de la classe RandomForestClassifier qui prend comme argument les ensembles les données explicatives `x_train` et les données cibles `y_train`.

````
##Importations

#importer la classe decisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

#importer la fonction train_test_split 
from sklearn.model_selection import train_test_split 

#importer scikit-learn metrics module pour le calcul de l'accuracy 
from sklearn import metrics 

#importer la fonction export_graphviz pour afficher l'arbre dans le notebook Jupyter
from sklearn.tree import export_graphviz

#importer pydotplus
import pydotplus

from sklearn.externals.six import StringIO  
from IPython.display import Image  

#importer la classe RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("/content/drive/MyDrive/ML/data/pima.csv", header=None, names=col_names)
pima.head()

#création d'une instance de la classe
rfc = RandomForestClassifier()

#model
rfc = RandomForestClassifier(random_state=0)

rfc.fit(x_train,y_train)

y_pred = rfc.predict(x_test)

#accuracy
print("Accuracy",rfc.score(x_test,y_test))





#exécution de l'instance sur les données d'apprentissage
#c.à d . construction du modèle prédictif
````

# Classificatin par la forêt d'arbres (suite)


````
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
````
